```python
import numpy as np


class CompressorPhysics:

    def __init__(self,
                 mass_flow,
                 pressure_ratio,
                 rpm,
                 Tin,
                 Pin,
                 efficiency,
                 blade_count):

        # Design inputs
        self.m_dot = mass_flow
        self.PR = pressure_ratio
        self.rpm = rpm
        self.Tin = Tin
        self.Pin = Pin
        self.eta = efficiency
        self.Z = blade_count

        # Gas properties (air)
        self.gamma = 1.4
        self.R = 287

        self.Cp = self.gamma * self.R / (self.gamma - 1)


    # --------------------------------------------------
    # Thermodynamic model
    # --------------------------------------------------

    def thermodynamics(self):

        self.rho = self.Pin / (self.R * self.Tin)

        self.deltaT = self.Tin * ((self.PR) ** ((self.gamma - 1) / self.gamma) - 1) / self.eta

        self.deltaH = self.Cp * self.deltaT

        return {
            "density": self.rho,
            "deltaT": self.deltaT,
            "deltaH": self.deltaH
        }


    # --------------------------------------------------
    # Impeller sizing
    # --------------------------------------------------

    def impeller_sizing(self):

        self.U2 = np.sqrt(self.deltaH / 0.7)

        self.D2 = (60 * self.U2) / (np.pi * self.rpm)

        self.r2 = self.D2 / 2

        self.D1 = 0.35 * self.D2
        self.r1 = self.D1 / 2

        return {
            "tip_speed": self.U2,
            "impeller_diameter": self.D2,
            "eye_diameter": self.D1
        }


    # --------------------------------------------------
    # Velocity triangles
    # --------------------------------------------------

    def velocity_triangles(self):

        phi = 0.06
        psi = 0.7

        self.Vm = phi * self.U2
        self.Vu2 = psi * self.U2

        self.beta1 = np.degrees(np.arctan(self.Vm / self.U2))

        self.beta2 = 60

        return {
            "Vm": self.Vm,
            "Vu2": self.Vu2,
            "beta1": self.beta1,
            "beta2": self.beta2
        }


    # --------------------------------------------------
    # Mach number
    # --------------------------------------------------

    def mach_number(self):

        a = np.sqrt(self.gamma * self.R * self.Tin)

        self.M_tip = self.U2 / a

        return self.M_tip


    # --------------------------------------------------
    # Performance summary
    # --------------------------------------------------

    def performance(self):

        self.thermodynamics()
        self.impeller_sizing()
        self.velocity_triangles()

        mach = self.mach_number()

        return {
            "mass_flow": self.m_dot,
            "pressure_ratio": self.PR,
            "rpm": self.rpm,
            "efficiency": self.eta,
            "tip_speed": self.U2,
            "impeller_diameter": self.D2,
            "eye_diameter": self.D1,
            "mach_tip": mach,
            "beta1": self.beta1,
            "beta2": self.beta2
        }
```

```python
%%writefile meridional_geometry.py

import numpy as np


class MeridionalGeometry:

    def __init__(self, r1, r2):

        self.r1 = r1
        self.r2 = r2


    # ---------------------------------------
    # Hub curve
    # ---------------------------------------

    def hub_curve(self):

        x = np.linspace(self.r1, self.r2, 50)

        z = -0.02*(x-self.r1)

        return x, z


    # ---------------------------------------
    # Shroud curve
    # ---------------------------------------

    def shroud_curve(self):

        x = np.linspace(self.r1, self.r2, 50)

        z = 0.03*np.sqrt((x-self.r1)/(self.r2-self.r1))

        return x, z


    # ---------------------------------------
    # Blade height distribution
    # ---------------------------------------

    def blade_height(self):

        hub_x, hub_z = self.hub_curve()
        shr_x, shr_z = self.shroud_curve()

        height = shr_z - hub_z

        return hub_x, height


    # ---------------------------------------
    # Full meridional data
    # ---------------------------------------

    def meridional_channel(self):

        hub_x, hub_z = self.hub_curve()
        shr_x, shr_z = self.shroud_curve()

        return {
            "hub_x": hub_x,
            "hub_z": hub_z,
            "shroud_x": shr_x,
            "shroud_z": shr_z
        }
```

```python
import matplotlib.pyplot as plt
from meridional_geometry import MeridionalGeometry
from compressor_physics import CompressorPhysics

compressor = CompressorPhysics(1.56,2.6,44000,288.15,101325,0.92,12)
compressor.performance()

geom = MeridionalGeometry(compressor.r1,compressor.r2)

channel = geom.meridional_channel()

plt.plot(channel["hub_x"],channel["hub_z"],label="Hub")
plt.plot(channel["shroud_x"],channel["shroud_z"],label="Shroud")

plt.xlabel("Radius")
plt.ylabel("Axial direction")
plt.title("Meridional Flow Path")

plt.legend()
plt.show()
```

```python
%%writefile blade_meanline.py
import numpy as np

class BladeMeanline:

    def __init__(self, r1, r2, beta1, beta2):
        self.r1 = r1
        self.r2 = r2
        self.beta1 = beta1
        self.beta2 = beta2


    def logarithmic_spiral(self, points=200):

        wrap_angle = np.radians(55)

        theta = np.linspace(0, wrap_angle, points)

        k = np.log(self.r2 / self.r1) / wrap_angle

        r = self.r1 * np.exp(k * theta)

        x = r * np.cos(theta)
        y = r * np.sin(theta)

        # rotate blade so inlet aligns nicely
        rot = -wrap_angle / 2

        xr = x*np.cos(rot) - y*np.sin(rot)
        yr = x*np.sin(rot) + y*np.cos(rot)

        return xr, yr


    def nurbs_blade(self):

        t = np.linspace(0,1,100)

        P0 = np.array([self.r1,0])
        P1 = np.array([self.r1*1.2, self.r1*0.4])
        P2 = np.array([self.r2*0.8, self.r2*0.8])
        P3 = np.array([self.r2, self.r2*0.2])

        B0 = ((1-t)**3)[:,None]
        B1 = (3*(1-t)**2*t)[:,None]
        B2 = (3*(1-t)*t**2)[:,None]
        B3 = (t**3)[:,None]

        curve = B0*P0 + B1*P1 + B2*P2 + B3*P3

        x = curve[:,0]
        y = curve[:,1]

        return x, y
```

```python
%%writefile cad_generator.py
import cadquery as cq


class CADGenerator:

    def __init__(self, r1, r2, blade_x, blade_y, blade_count):

        self.r1 = r1
        self.r2 = r2
        self.x = blade_x
        self.y = blade_y
        self.Z = blade_count

        self.hub_height = 0.01
        self.blade_height = (r2 - r1) * 0.5


    def hub(self):

        hub = (
            cq.Workplane("XY")
            .circle(self.r1)
            .extrude(self.hub_height)
        )

        return hub


    def shroud(self):

        shroud = (
            cq.Workplane("XY")
            .circle(self.r2)
            .extrude(0.002)
            .translate((0,0,self.blade_height))
        )

        return shroud


    def blade_surface(self):

        pts = list(zip(self.x, self.y))

        blade = (
            cq.Workplane("XY")
            .polyline(pts)
            .offset2D(0.002)
            .extrude(self.blade_height)
        )

        return blade


    def splitter_surface(self):

        start = int(len(self.x)*0.4)

        pts = list(zip(self.x[start:], self.y[start:]))

        splitter = (
            cq.Workplane("XY")
            .polyline(pts)
            .offset2D(0.002)
            .extrude(self.blade_height)
        )

        return splitter


    def blade_array(self):

        blade = self.blade_surface()
        splitter = self.splitter_surface()

        assembly = cq.Workplane("XY")

        pitch = 360/self.Z
        shift = pitch/2

        for i in range(self.Z):

            angle = pitch*i

            assembly = assembly.union(
                blade.rotate((0,0,0),(0,0,1),angle)
            )

            assembly = assembly.union(
                splitter.rotate((0,0,0),(0,0,1),angle+shift)
            )

        return assembly


    def build_impeller(self):

        hub = self.hub()
        blades = self.blade_array()
        shroud = self.shroud()

        impeller = hub.union(blades).union(shroud)

        return impeller


    def export_step(self, filename="impeller.step"):

        model = self.build_impeller()

        cq.exporters.export(model, filename)

        print("STEP file exported:", filename)
```
